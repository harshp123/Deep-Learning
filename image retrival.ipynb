```python
import graphlab
```

```python
train_image=graphlab.SFrame("image_train_data")
test_image=graphlab.SFrame("image_test_data")
```

```python
train_image.head()
```

# Train knn model for retriving deep features

```python
model=graphlab.nearest_neighbors.create(train_image,features=["deep_features"],label="id")
```

# using image rerival model with deep features

```python
cat=train_image[18:19]
```

```python
graphlab.canvas.set_target("ipynb")
cat["image"].show()
```

```python
model.query(cat)
```

```python
def get_images_from_ids(query_result):
    return train_image.filter_by(query_result["reference_label"],'id')
```

```python
cat_neighbors=get_images_from_ids(model.query(cat))
```

```python
cat_neighbors["image"].show()
```
